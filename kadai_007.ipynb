```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

# サンプルデータの読み込み
df = pd.read_csv('train.csv')
df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv('category.csv')
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

# カテゴリー列の要素のユニークな出現頻度をカウントする
category_counts = df['カテゴリー'].value_counts()

# 棒グラフ化
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('カテゴリー')
plt.ylabel('出現頻度')
plt.title('カテゴリーの出現頻度')
plt.xticks(rotation=90)
plt.show()

# 商品番号ごとにカテゴリー列の要素をグループにまとめ、サブ列を集計する
grouped = df.groupby('商品番号')['カテゴリー'].value_counts()

# 基本統計量を算出する
statistics = grouped.describe()
print(statistics)
```
