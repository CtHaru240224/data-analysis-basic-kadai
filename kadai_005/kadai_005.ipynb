```python

import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib

# FigureとAxesを作成し、レイアウトを自動調整する
fig, axs = plt.subplots(2, 3, tight_layout=True)

# 折れ線グラフ
y0 = [1, 2, -5, 2]
axs[0, 0].plot(y0)
axs[0, 0].set_title('折れ線グラフ')


# sin関数
x = np.linspace(0, 10, 100)
y = np.sin(x)
axs[0, 1].plot(x, y)
axs[0, 1].set_title('sin関数')

# ヒストグラム
data = np.random.randn(1000)
axs[0, 2].hist(data, bins=30)
axs[0, 2].set_title('ヒストグラム')

# 散布図
x = np.random.rand(100)
y = np.random.rand(100)
axs[1, 0].scatter(x, y)
axs[1, 0].set_title('散布図')

plt.show()

```
